```python
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3 import A2C
```

```python
import gym

# Create the environment
env = gym.make("ALE/MsPacman-ram-v5", render_mode='rgb_array')  

# required before you can step the environment
env.reset()

# sample action:
print("sample action:", env.action_space.sample())

# observation space shape:
print("observation space shape:", env.observation_space.shape)

# sample observation:
print("sample observation:", env.observation_space.sample())
```

```python
# Model creation and learning

model = A2C('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=100000)  
```

```python
model.save("models/a2c_mspacman")
```

```python
# Testing the results with 10 episodes

episodes = 10
frames = []

for ep in range(episodes):
    obs = env.reset()
    obs = obs[0]
    done = False
    while not done:
        action, _states = model.predict(obs)
        obs, rewards, done, tr, info = env.step(action)
        img = env.render()
        frames.append(img)
        print(rewards)
```

```python
# Now show the animation:
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
plt.rcParams['animation.embed_limit'] = 50  # For example, set it to 30 MB

# To get smooth animations
import matplotlib.animation as animation
mpl.rc('animation', html='jshtml')
def update_scene(num, frames, patch):
    # print(num)
    patch.set_data(frames[num])
    return patch,

def plot_animation(frames, repeat=False, interval=40):
    fig = plt.figure()
    patch = plt.imshow(frames[0])
    plt.axis('off')
    anim = animation.FuncAnimation(
        fig, update_scene, fargs=(frames, patch),
        frames=len(frames), repeat=repeat, interval=interval)
    plt.close()
    return anim
```

```python
plot_animation(frames)
```
